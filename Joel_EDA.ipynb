```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('cleaned_df.csv')
df
```

```python
day_mean_totals = df.groupby('WEEKDAY').TOTALS.mean().plot(kind='bar', title="Mean Totals ") # want to know how to plot this in mpl
```

```python
df.STATION.value_counts()
```

```python
top_20_stations = df.groupby('STATION').TOTALS.mean().sort_values(ascending=False).head(20)
top_20_stations.plot(kind='bar', title='Mean Daily Traffic Per Station (Top 20)')
```

```python
df.groupby("STATION").size()[top_20_stations.index].sort_values(ascending=False).plot(kind='bar', title="Total Observation Count Per Station (Top 20)")
```
